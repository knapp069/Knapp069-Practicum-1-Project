```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage to scrape
url = "https://docs.logrhythm.com/lrsiem/7.15.0/web-console-user-guide"

# Send a GET request to the URL
response = requests.get(url)

# Parse the content if the request is successful
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract and print the title of the webpage
    print("Title of the webpage:")
    print(soup.title.string if soup.title else "No title found")

    # Extract and print content from h1 tags
    print("\nContent from <h1> tags:")
    for h1 in soup.find_all('h1'):
        print(h1.get_text(strip=True))

    # Extract and print content from h2 tags
    print("\nContent from <h2> tags:")
    for h2 in soup.find_all('h2'):
        print(h2.get_text(strip=True))

    # Extract and print content from h3 tags
    print("\nContent from <h3> tags:")
    for h3 in soup.find_all('h3'):
        print(h3.get_text(strip=True))
    
    # Extract and print content from p tags
    print("\nContent from <p> tags:")
    for p in soup.find_all('p'):
        print(p.get_text(strip=True))

    # Extract and print content from li tags
    print("\nContent from <li> tags:")
    for li in soup.find_all('li'):
        print(li.get_text(strip=True))

    # Extract and print content from table tags
    print("\nContent from <table> tags:")
    for table in soup.find_all('table'):
        for row in table.find_all('tr'):
            row_data = [cell.get_text(strip=True) for cell in row.find_all(['th', 'td'])]
            print(', '.join(row_data))
else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")
```
